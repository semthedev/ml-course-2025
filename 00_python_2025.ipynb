```python
# Python - почти практически псевдокод, можно не думать про всякие инженерные штуки


def quicksort(arr):
    if len(arr) <= 1:
        return arr
    pivot = arr[len(arr) // 2]
    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]
    return quicksort(left) + middle + quicksort(right)


print(quicksort([3, 6, 8, 10, 1, 2, 1]))
```

```python
def factorial(n: int):
    if n == 0:
        return 1
    else:
        return factorial(n - 1) * n
```

```python
factorial(3)
```

```python
def factorial_loop(n):
    accumulator = 1
    for i in range(1, n + 1):
        accumulator *= i
    return accumulator
```

```python
factorial_loop(3)
```

```python
from functools import reduce


def f(x, y):
    return x * y


def factorial_fold(n):
    return reduce(f, range(1, n + 1))
```

```python
factorial_fold(3)
```

```python
%%time
factorial_fold(300)
```

```python
%%time
factorial(300)
```

```python
%%time
factorial_loop(300)
```

```python
class Circle(object):
    pi = 3.14

    def __init__(self, radius=1):
        self.radius = radius

    def area(self):
        return self.radius * self.radius * Circle.pi

    # вот так делать не надо, для этого в python есть @property
    def setRadius(self, radius):
        self.radius = radius

    def getRadius(self):
        return self.radius


c = Circle(1)

c.setRadius(2)
print("Radius is: ", c.getRadius())
print("Area is: ", c.area())
```

```python
# кроме __init__ есть и другие специальные имена.


class Circle(object):
    pi = 3.14

    def __init__(self, radius=1):
        self._radius = radius

    def __del__(self):
        print("Circle deleted")

    def __str__(self):
        return f"I am a cyrcle of radius {self._radius}"

    def __hash__(self):
        return id(self)

    def area(self):
        return self._radius * self._radius * Circle.pi

    @property
    def radius(self):
        print("Get radius")
        return self._radius

    @radius.setter
    def radius(self, value):
        print("Set radius")
        self._radius = value

    @radius.deleter
    def radius(self):
        print("Delete radius")
        del self._radius
```

```python
class Animal:  # наследоваться от object не обязательно
    def __init__(self):
        print("Animal created")

    def whoAmI(self):
        print("Animal")

    def eat(self):
        print("Eating")


class Dog(Animal):
    def __init__(self):
        Animal.__init__(self)
        print("Dog created")

    def whoAmI(self):
        print("Dog")

    def bark(self):
        print("Woof!")


d = Dog()
```

```python
d.whoAmI()
```

```python
d.eat()
```

```python
d.bark()
```
